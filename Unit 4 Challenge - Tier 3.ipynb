**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose the DataFrame to swap rows with columns
properties_T = properties.T

# Reset the index to move the row labels into a column
properties_T = properties_T.reset_index()

# Set the first row as the header by assigning it as the new column names
properties_T.columns = properties_T.iloc[0]

# Drop the first row, which is now redundant since it was used as the header
properties_T = properties_T.drop(0)

# Display the first few rows to confirm the transformations
properties_T.head()
```

**2.4.Transforming the data**

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Creating a function to filter datframe for two decades which is between 1998 and 2018
def create_price_ratio(df):
    '''The create_price_ratio function is designed to 
    calculate the ratio of house prices between two specific years,
    1998 and 2018, for a given London borough. 
    This ratio helps determine how much house prices have increased over this 20-year period. '''
    
    # Get the average price for 1998
    price_1998 = df[df['Year'] == 1998]['Average_price'].mean()
    
    # Get the average price for 2018
    price_2018 = df[df['Year'] == 2018]['Average_price'].mean()
    
    # Calculate the ratio of 1998 price to 2018 price
    ratio = price_2018 / price_1998
    
    return ratio
```

```python
# Get the unique London Boroughs
boroughs = properties_df['London_Borough'].unique()

# Create a dictionary to store the results
price_ratios = {}

# Iterate through each borough and calculate the price ratio
for borough in boroughs:
    borough_data = properties_df[properties_df['London_Borough'] == borough]
    price_ratios[borough] = create_price_ratio(borough_data)

# Convert the dictionary to a DataFrame for better visualization
price_ratios_df = pd.DataFrame(list(price_ratios.items()), columns=['London_Borough', 'Price Ratio'])

# Display the DataFrame
price_ratios_df.sort_values(by='Price Ratio', ascending=False, inplace=True)
price_ratios_df.head()
```

```python

# Set up the figure and axis with a larger size
plt.figure(figsize=(30, 8))  # Increase the size of the plot

# Plot the price ratios as a bar chart
price_ratios_df.plot(kind='bar', x='London_Borough',  legend=False)

# Add a title to the plot
plt.title("Price Increase Across Each Borough in London Over Two Decades", fontsize=16)

# Label the x-axis and y-axis
plt.xlabel("London Borough", fontsize=14)
plt.ylabel("Price Ratio (2018/1998)", fontsize=14)


# Display the plot
plt.tight_layout()  # Adjusts the plot to ensure everything fits without overlapping
plt.show()
```
